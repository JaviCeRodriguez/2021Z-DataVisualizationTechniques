## Wykres poprawiony

```python
import matplotlib.pyplot as plt
import numpy as np
labels = ["2017 Q4", "2018 Q1", "2018 Q2", "2018 Q3", "2018 Q4", "2019 Q1", "2019 Q2", "2019 Q3", "2019 Q4", "2020 Q1", "2020 Q2", "2020 Q3"]
values = [866.35, 1092.09, 886.34, 697.23, 827.97, 1039.15, 928.07, 689.78, 846.1, 1092.85, 712.09, 626.74]

fig, ax = plt.subplots(figsize=(15,9))
plt.bar(labels, values, zorder=2)
plt.grid(zorder=0, axis="y")

plt.title("GDP From Mining in India", fontsize=26)
plt.ylabel("GDP, in billions of INR", fontsize=16)

for i, v in enumerate(values):
    ax.text(i, v+12, str(v), ha='center', fontsize=14)

plt.show()
```

## Wprowadzone zmiany
W wykresie została zmieniona oś pionowa, tak aby zaczynała się w zerze. Wartości zaznaczone na osi poziomej zostały zmienione z wybranych miesięcy na kwartały, aby lepiej odzwierciedlały prezentowane dane. Do wykresu został także dodany tytuł, a oś pionowa została podpisana, z uwzględnieniem jednostki w jakiej przedstawiane są dane.
