# Wykres Poprawiony

```python
d={"Atr":["Mwm","Rigour","Gg","T"],
   "THE":[32,30,50,45],
   "ARWU":[25,43,38,79],
   "LEIDEN":[66,87,62,79],
   "Multirank":[75,38,66,75],
   "QS":[30,38,72,58],
   "US NEWS":[28,42,42,58]}
df=pd.DataFrame(d)
pos=list(range(len(df['THE'])))
width=0.15
fig,ax=plt.subplots(figsize=(10,5))
plt.grid(axis='y')
plt.bar(pos, df['THE'], width, color='#377eb8')
plt.bar([p + width for p in pos], df['ARWU'], width, color='#e41a1c')
plt.bar([p + width*2 for p in pos], df['LEIDEN'], width, color='#999999')
plt.bar([p + width*3 for p in pos], df['Multirank'], width, color='#4daf4a')
plt.bar([p + width*4 for p in pos], df['QS'], width, color='#984ea3')
plt.bar([p + width*5 for p in pos], df['US NEWS'], width, color='#ff7f00')
ax.set_xticks([p + 2.5 * width for p in pos])
ax.set_xticklabels(['Measure what matters', 'Rigour', 'Good governance', 'Transparency'])
plt.legend(["THE","ARWU","LEIDEN","Multirank","QS","US NEWS"], loc='upper left',bbox_to_anchor=(1, 1),title="World Rankers")
plt.title("Total possible score for each world ranker.")
plt.ylabel('Score in %')
plt.show()
```

# Komentarz
Mój wykres jest lepszy ponieważ jest prostym zgrupowanym barchartem, na którym łatwo porównać jak oceniane są konkretne cechy każdego ranker'a. Dodatkowo nie występuje żaden z problemów oryginalnego wykresu np. nic na siebie nie nachodzi, kolejność cech nie ma wpływu na odczyt wykresu (na radarze ma).
