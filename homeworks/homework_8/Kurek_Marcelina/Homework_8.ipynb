Wykres przedstawia media społecznościowe, w których ludzie doświadczają cyberprzemocy.

## Poprawiony wykres

```python
import matplotlib.pyplot as plt
nazwy = ['Twitter', 'YouTube', 'WhatsApp', 'Snapchat', 'Facebook', 'Instagram']
wartosci = [9, 10, 12, 31, 37, 42]
procent = ["9%", "10%", "12%", "31%", "37%", "42%"]
```

```python
fig = plt.figure(figsize = (8, 7))
plt.bar(nazwy, wartosci)

for i, v in enumerate(wartosci):
    plt.text(i -0.15, (wartosci[i]+ 0.6), str(v), color='black')
    plt.text(i  + 0.1, wartosci[i]+ 0.6, "%", color='black')

plt.title("Where are people cyberbullied?")

plt.gca().axes.yaxis.set_visible(False)
plt.gca().axes.set_ylim([0, 45])
   
```

## Poniższy wykres zawiera następujące poprawione elementy:
- zamiana słupków 3D na 2D
- ujednolicenie koloru
- usunięcie osi 0Y (wartości znajdują się bezpośrednio nad słupkami)
- poziome napisy zamiast napisów pod kątem
- brak symboli graficznych
