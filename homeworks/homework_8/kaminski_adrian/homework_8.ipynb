## Wykres poprawiony

```python
labels = ['NL', 'SK','AB','MB','QC','ON','NB','NS','PE','BC'] # szacowania na podstawie wykresu 
val2020 = [-9,-8,-7.5,-6,-5.8,-5.6,-5.5,-5.2, -5, -4.8]  # z ktorego jednak ciezko cokolwiek odczytac
val2021 = [3, 1.8, 1.5,1.2,1.4,1.3,1.9, 1.7,0.95,1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    "text.usetex": True
})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey='row',figsize=(16,9),
                              gridspec_kw={'hspace': 0, 'wspace': 0})



fig.suptitle("\huge{Real GDP in 2020/2021}\n \large{Percent change}" , va="top", ha='left',x=.13)
ax1.bar(labels, val2020, color = 'red')
ax1.set_title('2020', size=16)
ax1.set_ylabel('Change in \%', size=15)
ax2.bar(labels, val2021, color = 'aqua')
ax2.set_title('2021', size=16)

ax1.grid(axis="y", alpha=0.5)
ax2.grid(axis="y", alpha=0.5)

ax1.set_axisbelow(True)
ax2.set_axisbelow(True)

plt.show()
```
