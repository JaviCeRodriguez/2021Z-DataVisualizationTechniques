# Wykres poprawiony

```python
tablica = {'party': ["Reformierakond", "Eesti 200", "Keskerakond", "Ekre"], 'percentage': [0.28, 0.18, 0.17, 0.16]}
df = pd.DataFrame(data = tablica)
plt.figure(figsize = (13,10))
sns.set(font_scale = 1.6)
sns.set_style("white")
pl = sns.barplot(data = df, x = 'party', y = 'percentage', color = 'grey')
pl.set(xlabel = '', ylabel = '')
ticks = pl.get_yticks()
pl.set_yticklabels(['{:,.0%}'.format(x) for x in ticks])
plt.title("Wykres popularności estońskich partii", size = 22)
plt.show()
```

# Uzasadnienie

Pierwotny obraz ciężko właściwie nazwać wykresem. Są to raczej cztery prostokąty umieszczone jeden pod drugim bez ładu i składu. Wziąłem więc zapisane na nich wartości, i zamieniłem w prostu wykres słupkowy. Choć sama poprawa nie jest zbyt odkrywcza, to sam "wykres" oryginalny uznałem za bardzo ciekawy przykład manipulacji wykresami w mediach - widać tutaj wyraźnie stronniczość co do niektórych partii. Procenty oczywiście nie sumują się do 100, ale uznałem że na tym wykresie chciano raczej pokazać różnice między tymi czterema konretnymi partiami.
