## Poprawiony wykres

```python
import matplotlib.pyplot as plt
import numpy as np

procent = [78, 13, 5, 4]
opis = ['desaprueba', 'aprueba', 'no aprueba \nni desaprueba', 'no sabe \nde responder']

fig = plt.figure()
plt.bar(opis, procent, color="#3399ff")
plt.title('Aprobación del Presidente de Chile', fontsize=20)
plt.xticks(fontsize=14)
plt.ylabel('El porcentaje', fontsize=16)
plt.yticks(fontsize=14)
plt.rc('axes', axisbelow=True)
plt.grid(axis='y')
fig.set_figheight(7)
fig.set_figwidth(12)

plt.show()
```

## Dokonane zmiany

Poprawiany wykres jest wykresem kołowym, w którym zaznaczone wycinki koła są nieproporcjonalne do podpisów. Wycinek koła odzwierciedlający 13% poparcie dla urzędującego prezydenta wygląda na 25%. Z drugiej strony procent osób niepopierający urzędującego prezydenta (78%) wygląda na nieznacznie ponad 50%. Dlatego też w przedstawionej widzom formie, wykres ten manipuluje danymi. Zatem w celu rzetelnego przedstawienia tych informacji zmieniłem typ wykresu na słupkowy i zastosowałem poprawną skalę.
