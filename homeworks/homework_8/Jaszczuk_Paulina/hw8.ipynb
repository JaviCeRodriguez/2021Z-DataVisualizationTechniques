## Wykres poprawiony przeze mnie:

#### Import 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.notebook_repr_html", False)
```

#### Praca z danymi

```python
regions = np.array(["Dalmacija", "Istra i Primorje sa zaledem", "Zagreb", "Sjeverna Hrvatska", "Središnja Hrvatska", "Slavonija"])
pro_maskers = np.array([91.90, 84.80, 87.20, 84.30, 84.50, 87.80])
anti_maskers = np.array([8.10, 15.20, 12.80, 15.70, 15.50, 12.20])

regions = np.sort(regions)
```

#### Tworzenie wykresu

```python
ax0 = plt.barh(regions, pro_maskers, align='center', height=.70, color='#30DCFF',label='pro-maskers')

for rec in ax0.patches:
    width = rec.get_width()
    plt.text(rec.get_x() + rec.get_width() / 2, 
              rec.get_y() + height / 2  - 0.1,
              "{:.1f}%".format(width),
              ha='center', 
              va='bottom')

ax1 = plt.barh(regions, anti_maskers, align='center', height=.70, left=pro_maskers, color='#F91955',label='anti-maskers')

for rec in ax1.patches:
    width = rec.get_width()
    plt.text(rec.get_x() + rec.get_width() / 2, 
              rec.get_y() + height / 2 - 0.1,
              "{:.1f}%".format(width),
              ha='center', 
              va='bottom')

plt.ylabel('region', size = 13)
plt.xlabel('percent', size = 13)
plt.legend(loc = "lower center", bbox_to_anchor=(0.5, -0.3), ncol = 2)
plt.title("Ratio between pro-maskers and anti-maskers in Croatia by region", loc = "center", size = 20)
plt.show()
```
