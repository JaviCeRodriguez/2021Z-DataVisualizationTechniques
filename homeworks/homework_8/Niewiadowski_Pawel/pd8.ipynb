```python
names = ['Sjeverna Hrvatska','Sredisnja Hrvatska','Istra i Primorje sa zaledem','Zagreb','Slavonija', 'Dalmacija']
values = [15.70,15.50,15.20,12.80,12.20,8.10]
```

```python
plt.barh(names, values)
plt.xlabel('%')
plt.title("Odsetek nienoszących masek z podziałem na region Chorwacji")
plt.xlim(0, 20) 
for index, value in enumerate(values):
    plt.text(value - 3, index -.1, str(value) + " %", color="white")
plt.show()
```
