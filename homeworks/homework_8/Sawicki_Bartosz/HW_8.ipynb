## Poprawiony wykres

```python
vals = [38,26,13,11,7,5]
parties = ["PiS","PO","2050","Lewica","PSL","Konfederacja"]
colors = ["#000099","#ff751a","#ffd633","#ff0000","#66ffb3","#000066"]

sns.set(rc={'figure.figsize':(15,10)})
sns.set_style("whitegrid")

f = sns.barplot(y = vals, x = parties, palette= colors)

f.set_ylabel("Poparcie w %", size = 20)
plt.xticks(size = 16)
plt.yticks(size = 16)
plt.title("Poparcie dla partii politycznych w listopadzie 2020", size = 30)

plt.show()
```

## Co i dlaczego zmieniono?

 - tytuł nie był jasny, co w połączeniu z niepodpisaną osią Y mogło prowadzić do nieporozumień. Nie było jasne czy poparcie wzrosło o 38% czy wyniosło 38%
 - strzałki na najwyższym słupku były niepotrzebne
